```python
from ultralytics import YOLO

# Load your trained model
model = YOLO('/content/runs/detect/accident_detection_v1/weights/best.pt')

def predict_accident(source):
    results = model(source, conf=0.25)
    for r in results:
        if len(r.boxes) > 0:
            print("🚨 Accident detected!")
            return r.plot()  # Return annotated image
        else:
            print("✅ No accident")
            return r.orig_img

# Example usage
result = predict_accident("/content/drive/MyDrive/Examples/Acc2.png")
```

```python
from ultralytics import YOLO
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Image

# 1. Load your trained model
model = YOLO('/content/runs/detect/accident_detection_v1/weights/best.pt')

# 2. Run Validation with Correct Data Paths
metrics = model.val(
    data='/content/Accident-detection-model-2/data.yaml',  # Make sure this points to your correct YAML
    split='test',
    conf=0.25,
    iou=0.50,
    plots=True
)

# 3. Extract and Format Metrics (with NumPy conversion)
def safe_metric(metric):
    return float(metric[0]) if isinstance(metric, np.ndarray) else metric

metrics_dict = {
    'Precision': safe_metric(metrics.box.p),
    'Recall': safe_metric(metrics.box.r),
    'mAP50': safe_metric(metrics.box.map50),
    'mAP50-95': safe_metric(metrics.box.map)
}

# 4. Display Metrics
print("🚗 Accident Detection Model Performance 🚗")
print(f"📌 Precision: {metrics_dict['Precision']:.3f} (Correct accident predictions)")
print(f"📌 Recall: {metrics_dict['Recall']:.3f} (Actual accidents detected)")
print(f"📌 mAP50: {metrics_dict['mAP50']:.3f} (Overall detection quality)")
print(f"📌 mAP50-95: {metrics_dict['mAP50-95']:.3f} (Localization accuracy)")

# 5. Visualization
plt.figure(figsize=(10, 5))
plt.bar(metrics_dict.keys(), metrics_dict.values(), color=['blue', 'green', 'orange', 'red'])
plt.title('Model Performance Metrics', fontweight='bold')
plt.ylim(0, 1)
plt.ylabel('Score')
plt.grid(axis='y', alpha=0.3)
plt.show()

# 6. Show Generated Plots
print("\n🔍 Detailed Analysis Plots:")
display(Image(filename='/content/runs/detect/val2/confusion_matrix.png'))
display(Image(filename='/content/runs/detect/val2/F1_curve.png'))

# 7. Class-Specific Performance (Since you have nc=1)
if hasattr(metrics, 'speed'):
    print(f"\n⚡ Inference Speed: {metrics.speed['inference']:.1f}ms/img")

# 8. Traditional Accuracy Calculation
try:
    tp = safe_metric(metrics.box.tp)
    fp = safe_metric(metrics.box.fp)
    fn = safe_metric(metrics.box.fn)
    accuracy = tp / (tp + fp + fn) if (tp + fp + fn) > 0 else 0
    print(f"\n🎯 Traditional Accuracy: {accuracy:.3f}")
except AttributeError:
    print("\nℹ️ Traditional accuracy not available in this validation mode")
```
